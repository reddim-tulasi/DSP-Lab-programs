```python
import matplotlib.pyplot as plt
country=['USA','India','Germany','Uk','France']
GDP=[48000,38000,50000,45000,44000]
plt.bar(country,GDP)
plt.title('country vs GDP')
plt.xlabel('country')
plt.ylabel('GDP')
plt.show()
```

```python
import matplotlib.pyplot as plt
numbers=[0.1,0.5,1,1.5,2,4,5.5,8,9]
plt.hist(numbers,bins=5)
plt.xlabel=("Numbers")
plt.ylabel=("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
co=("r","b","hotpink")
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.legend()
plt.show()
```
