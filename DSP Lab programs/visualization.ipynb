```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints=np.array([0,200])
ypoints=np.array([0,1000])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y,'g',linewidth=6,linestyle="dashdot")
plt.show()
```

```python
plt.plot(x,y,'g',linewidth=6,linestyle="-")
plt.show()
```

```python
plt.plot(x,y,'g',marker='<',markersize='20',markeredgecolor='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
departments=['cse','civil','mec','ece']
students=[100,200,300,400]
plt.title("rgukt")
plt.xlabel("departments")
plt.ylabel("students")
plt.bar(departments,students)
plt.show()
```

```python
plt.bar(departments,students,width=0.3,color="r")
```

```python
plt.bar(departments,students,width=[0.2,0.5,0.3,0.7],color="y")
```

```python
plt.bar(departments,students,width=0.3,color=['r','b','g','y'])
plt.show()
```
