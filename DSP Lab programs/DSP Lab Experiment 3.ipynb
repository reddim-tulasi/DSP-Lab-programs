```python
#visualization graph plotting
#line graph
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,7])
y=np.array([2,4])
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.show()
```

```python
#making lines with given input marker style
x=np.array([1,7])
y=np.array([2,4])
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,"--",color="black")
plt.show()
```

```python
#using markers
x=np.array([1,7,8,9])
y=np.array([6,2,8,4])
plt.plot(x,y,marker="o",mec="black")
plt.show()
```

```python
#giving grid
x=np.array([2,3])
y=np.array([4,6])
plt.plot(x,y)
plt.grid()
```

```python
#tow plots in a single graph
x=[1,2,7,9]
y=[8,2,3,0]
plt.plot(x,y,color="black",marker="o",mec="blue")
x=[9,10,1,9]
y=[10,3,4,6]
plt.plot(x,y,color="blue",marker="o",mec="blue")
```

```python
#bar graphs
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,4,6])
y=np.array([7,8,1,0])
plt.bar(x,y)
```

```python
#horizontal bar graphs
x=np.array([1,2,4,6])
y=np.array([7,8,1,0])
plt.barh(x,y)
```

```python
#giving color to the bars
x=np.array([1,3,4,6])
y=np.array([7,8,9,0])
plt.bar(x,y,color="blue")
```

```python
#title
x=np.array([3,4,8,7])
y=np.array([6,4,8,6])
col=["red","blue","pink","yellow"]
plt.bar(x,y,color=col)
plt.title("bar graph")
plt.show()
```

```python
#Labels
x=["a","b","c","d"]
y=[10,8,6,4]
plt.xlabel("alphabets")
plt.ylabel("their values")
plt.title("Bar Graphs")
plt.bar(x,y)
plt.show()
```

```python
#Pie charts
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

```python
#mentioning the labels for each divisions
y=np.array([35,15,50])
plt.pie(y,labels=["A","B","C"])
plt.legend()
plt.show()
```

```python
#different colors to different divisions
y=np.array([35,15,50])
co= ("pink","blue","yellow")
plt.pie(y,colors=co)
plt.show()
```

```python
#labelled to the divisions and colors
y=np.array([35,50,15])
plt.pie(y,labels=["A","B","C"])
plt.legend(title="sum")
plt.show()
```

```python
#to separate the selected division
y=np.array([35,50,15])
plt.pie(y,explode=[0,0.2,0])
plt.show()
```

```python
#scatter plot
x=[2,7,8,9,1]
y=[9,2,7,1,8]
plt.scatter(x,y)
plt.show()
```

```python
#different colors to different points
x=[2,7,8,9,1]
y=[9,2,7,18,7]
c=["red","blue","green","violet","yellow"]
size=[10,20,30,40,50]
plt.scatter(x,y,color=c,s=size)
plt.show()
```

```python
#different markers,sizes,colors to the points
x=[2,7,8,9]
y=[9,2,1,2]
c=["red","blue","green","orange"]
size=[10,20,30,40]
plt.scatter(x,y,color=c,s=size,marker=">")
plt.show()
```

```python
#square,edge color,line width
x=[2,7,8,9]
y=[9,2,1,2]
c=["red","blue","green","violet"]
size=[10,20,30,40]
plt.scatter(x,y,color=c,s=size,marker="s",edgecolor="red",linewidth=2)
plt.show()
```

```python
x=[1,2,3,4]
y=[8,9,7,6]
plt.scatter(x,y,s=10)
plt.show()
```

```python
#different colors for different points
x=[1,2,3,4]
y=[8,9,7,6]
co=["red","blue","green","violet"]
z=[10,20,30,40]
plt.scatter(x,y,color=co,s=z,alpha=0.5)
plt.show()
```

```python
#random colors,values and sizes
x=np.random.rand(40)
y=np.random.rand(40)
co=np.random.rand(40)
z=np.random.rand(40)
plt.scatter(x,y,c=co,s=z*100,marker="s",edgecolor="red",linewidth=1)
plt.show()
```

```python
#box plot
x=[1,2,3,7,8,9]
plt.boxplot(x)
```

```python
#from the data frame
import pandas as pd
df=pd.DataFrame(np.random.rand(15,4)*10,columns=["a","b","c","d"])
df.boxplot()
```

```python
#histogram
n=[0.1,0.5,1,1.5,2.4,5,5.5,6,8,9]
plt.hist(n,bins=5)
plt.xlabel("n")
plt.ylabel("frequency")
plt.show()
```

```python
#subplot
plt.subplot(1,2,1)
plt.plot([10,20,50,30],[5,10,15,6])
plt.subplot(1,2,2)
plt.plot([0,2,3,4],[4,1,6,7])
plt.show()
```
